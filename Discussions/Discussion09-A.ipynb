```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA

np.random.seed(1)


n_samples=1000
x,y = datasets.make_blobs(n_samples=n_samples, random_state=1, n_features=2, centers=3)
x1=x[:,0]
x2=x[:,1]

plt.scatter(x1,x2)
```

```python
pca = PCA(n_components=1)
pca.fit(x)
print(pca.explained_variance_ratio_)
```

```python
x_new=pca.transform(x)
plt.scatter(x_new, [0]*len(x_new))
```
